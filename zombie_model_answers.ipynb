# A mathematical model of a zombie apocalypse

A warm-up activity for UWE Maths and Stats students

## Problem description

Let's model a zombie apocalypse.

$n$ = the number of days since the start of the invasion

$S_n$ = the number of susceptible humans after $n$ days

$Z_n$ = the number of zombies after $n$ days

$R_n$ = the number of zombies killed by humans after $n$ days

$a$ = the proportion of humans that a single zombie will bite each day

$b$ = the proportion of zombie that a human will kill each day

$$S_{n+1}=S_n-aS_n Z_n$$

$$Z_{n+1}=Z_n+(a-b) S_n Z_n$$

$$R_{n+1}=R_n+bS_n Z_n$$

Who will survive?


## Creating a simulation step by step

Let's create our variables, first the parameters of prograpagation of teh zombie epidemy $a$ and $b$

```python
a=0.0095
```

```python
b=0.005
```

Then the lengths of time for the model to run $N$ (in days)

```python
N=100
```

Now let's create a variable time $T$, counting the number of days since the begining of the simulation

```python
T=range(N)
```

$S$, $N$ and $R$ are vectors of the number of human, zombie and dead at each day. We will store them in table of size $N$, with $S[i]$ ($N[i]$, $R[i]$) the number of humans (zombie, dead) at time $i$

```python
S=[0]*100
```

```python
Z=[0]*100
```

```python
R=[0]*100
```

Finally let's set our initial populations sizes, 199 humans and 1 zombie

```python
S[0]=199
```

```python
Z[0]=1
```

```python
R[0]=0
```

The first day:

```python
S[1]=S[0]-a*S[0]*Z[0]
Z[1]=Z[0]+(a-b)*S[0]*Z[0]
R[1]=R[0]+b*S[0]*Z[0]
```

```python
(S[1],Z[1],R[1])
```

Let's write this for all time $i$ until $N$

```python
for i in range(1,N):
    S[i]=S[i-1]-(a*S[i-1]*Z[i-1])
    Z[i]=Z[i-1]+((a-b)*S[i-1]*Z[i-1])
    R[i]=R[i-1]+(b*S[i-1]*Z[i-1])
```

```python
S[:10]
```

```python
Z[:10]
```

```python
R[:10]
```

Now let's plot this:

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(T,S,color="blue",label="humans")
plt.plot(T,Z,color="red",label="zombies")
plt.plot(T,R,color="black", label="dead")
plt.legend()
plt.xlabel("time in days")
plt.ylabel("number of individuals")
```

## Generic simulation

Now let's put it all in one place to do different simulations

```python
a=0.0095
b=0.005
N=20 # here we put 20 since all humans were already dead at 20
T=range(N)
S=[0]*N
Z=[0]*N
R=[0]*N
S[0]=200
Z[0]=1
R[0]=0
```

```python
for i in range(1,N):
    S[i]=S[i-1]-(a*S[i-1]*Z[i-1])
    Z[i]=Z[i-1]+((a-b)*S[i-1]*Z[i-1])
    R[i]=R[i-1]+(b*S[i-1]*Z[i-1])
plt.plot(T,S,color="blue",label="humans")
plt.plot(T,Z,color="red",label="zombies")
plt.plot(T,R,color="black", label="dead")
plt.legend()
plt.xlabel("time in days")
plt.ylabel("number of individuals")
```
